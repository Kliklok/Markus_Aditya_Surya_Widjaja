```python
circuits = pd.read_csv("D:/Mini Project/circuits.csv")
display(circuits)
```

```python
races = pd.read_csv("D:/Mini Project/races.csv")
races.tail(19)
```

```python
# Filtering for race year
r2014=races[races['year']>2019]
r2014.shape
```

```python
constructor = pd.read_csv("D:/Mini Project/constructors.csv")
constructor['constructorId'].astype(int)
```

```python
rc = races.merge(circuits[['circuitId','name']].set_index('circuitId'), on='circuitId')
display(rc)
```

```python
rcr=results.merge(rc[['circuitId','raceId','name_y']].set_index('raceId'), on='raceId')
rcr
```

```python
rcr['position'].astype(str)

rcr.dtypes
```

```python
rcragg=rcr[rcr['position']=="1"].groupby(['circuitId','constructorId'])[['raceId']].count().reset_index()
rcraggmax=rcragg[['circuitId','constructorId','raceId']].groupby(['circuitId'])[['raceId']].max()
rcragg
```

```python
rcraggmax
```

```python
merged_df = rcraggmax.merge(rcragg, how='inner', left_on=["name_y", "raceId"], right_on=["name_y","raceId"])
merged_df.rename(columns={'name_y':'Circuit_Name'},inplace=True)
merged_df.rename(columns={'raceId':'No.of_Win'},inplace=True)
merged_df
```

```python
c=merged_df.merge(constructor, how='inner', left_on=["constructorId"], right_on=["constructorId"])
c
```

```python
MostWin=c[["Circuit_Name","name","No.of_Win"]]
MostWin.loc[MostWin['Circuit_Name'] == 'Circuit Paul Ricard']
```

```python
# MostWin table showing the constructor with the highest number of win for each track for 2020-2021 season (up to Styrian GP)
# realise that uploading this to PowerBI won't allow for any further race year filter as the aggregated no. of wins is already
# detatched from each respective raceID and race year
MostWin
```

```python
e=r2014.merge(circuits,how='inner', left_on=['circuitId'],right_on=['circuitId'])
r2014_merge=e[['circuitId','name_y','year','name_x']]
Circuits_Hybridera=r2014_merge['name_y'].unique()
Circuits_Hybridera
```

```python
g=MostWin.merge(df,how='inner', left_on=['Circuit_Name'],right_on=['circuits'])
h=g[['Circuit_Name','name','No.of_Win']]
h.rename(columns={'name':'Constructor'},inplace=True)
h
```

```python
# Making a table with raceID, driverId, and cumulative number of overtakes
Laptimes = pd.read_csv("D:/Mini Project/lap_times.csv")
Laptimes
```

```python
Laptimes.tail(20)
```

```python
Laptimes['lag'] = Laptimes.sort_values(by=['lap'], ascending=True).groupby(['raceId','driverId'])['position'].shift(1)
Laptimes.head(60)
```

```python
import numpy as np
Laptimes['Overtake']= np.where(Laptimes['lag']-Laptimes['position']>0,Laptimes['lag']-Laptimes['position'],0)

Laptimes.head(60)
```

```python
Laptimes[(Laptimes['raceId']==1) & (Laptimes['driverId']==1)]
```

```python
# Table with raceId, driverId, and cumulative number of overtakes
ot=Laptimes.groupby(['raceId','driverId'])[['Overtake']].sum().reset_index()
ot.head(20)
```

```python
ot[ot['raceId']==1]
```

```python
ot.groupby(['raceId'])[['Overtake']].sum()
```

```python
Laptimes['poschange']=Laptimes['lag']-Laptimes['position']
Laptimes.head(20)
```

```python
# Table with raceId, driverId, and cumulative number of position change from race start to finish
posfin=Laptimes.groupby(['raceId','driverId'])[['poschange']].sum().reset_index()
posfin.head(20)
```
